```r
# https://www.kaggle.com/c/tabular-playground-series-dec-2021/overview
#          
# 

library(tidyverse)
library(caret)
library(e1071)
library(class)
library(rpart)
library(ipred)
library(xgboost)
library(randomForest)

#  
#     readr   
data <- read_csv('./data/tabular-playground-series-dec-2021_train.csv')

str(data)
#summary(data)

#   
# colSums(is.na(data))
# 1  
set.seed(2021)
data.sampled <- data[sample(nrow(data), 10000), ]

#   
hist(data.sampled$Cover_Type)

#  
part <- createDataPartition(
    data.sampled$Cover_Type
    , p = 0.7
)

data.train <- data.sampled[part$Resample1, ]
data.test <- data.sampled[-part$Resample1, ]

nrow(data.train); nrow(data.test)

#     
# ,        
# model.lm <- lm(
#     data = data.train
#     , formula = Cover_Type ~ .
# )

# model.lm.step <- step(
#     model.lm
#     , direction = "both"
# )

# summary(model.lm.step)

# AIC          
# lm(formula = Cover_Type ~ Id + Elevation + Slope + Vertical_Distance_To_Hydrology + 
#     Hillshade_Noon + Hillshade_3pm + Horizontal_Distance_To_Fire_Points + 
#     Wilderness_Area3 + Wilderness_Area4 + Soil_Type2 + Soil_Type4 + 
#     Soil_Type6 + Soil_Type19 + Soil_Type21 + Soil_Type26 + Soil_Type28 + 
#     Soil_Type31 + Soil_Type33 + Soil_Type35 + Soil_Type37 + Soil_Type38 + 
#     Soil_Type39 + Soil_Type40, data = data.train)

# SVM
#      
model.svm <- svm(
    formula = Cover_Type ~ Id + Elevation + Slope + Vertical_Distance_To_Hydrology + 
        Hillshade_Noon + Hillshade_3pm + Horizontal_Distance_To_Fire_Points + 
        Wilderness_Area3 + Wilderness_Area4 + Soil_Type2 + Soil_Type4 + 
        Soil_Type6 + Soil_Type19 + Soil_Type21 + Soil_Type26 + Soil_Type28 + 
        Soil_Type31 + Soil_Type33 + Soil_Type35 + Soil_Type37 + Soil_Type38 + 
        Soil_Type39 + Soil_Type40
    , data = data.train
    , type = "C-classification"
)

model.svm.pred <- predict(
    model.svm
    , newdata = data.test
)

model.svm.pred
data.test$Cover_Type

confusionMatrix(
    data = as.factor(model.svm.pred)
    , reference = as.factor(data.test$Cover_Type)
)

# Accuracy : 0.89

#     
result.data.test <- read_csv('./data/tabular-playground-series-dec-2021_test.csv')

result.model.svm.pred <- predict(
    model.svm
    , newdata = result.data.test
)

result.df <- cbind(result.data.test$Id, result.model.svm.pred)
colnames(result.df) <- c("Id", "Cover_Type")

write.csv(result.df, './data/20211206-R.csv', row.names = FALSE)


# KNN

# 

# 

# 

# xgboost

# 




```
