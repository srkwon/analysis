```r

###################################################################################
#                      1. 통계분석 (사용 데이터 : Carseats)                  
###################################################################################

#------------------------------------------------------------------------------------------------
# Q1) Urban변수에 따른 Sales의 차이가 있는지를 통계적으로 검증하기 위한 통계분석을 수행하고, 결과를 해석하시오. 
#    (데이터는 정규성을 만족한다고 가정하고 유의수준 0.05 하에서 검정
#------------------------------------------------------------------------------------------------

library(ISLR)

# 데이터 수집 및 확인, 전처리

data(Carseats)
df <- Carseats

summary(df)
str(df)
head(df)

# Urban 변수 => 범주형(factor), Sales(수치형) 차이 확인
table(df$Urban)
hist(df$Sales)

# 등분산 검정 수행
# 귀무가설 : 두 집단의 분산이 동일하다.
# 대립가설 : 두 집단의 분산이 동일하지 않다.
var.test(Sales~Urban, data=df, alternative="two.sided") 

# 두 모집단이 등분산성 가정을 만족한다는 가정하에서 독립표본 t-검정을 수행 
# 귀무가설 : 차이가 없다
# 대립가설 : 차이가 있다.
# 유의수준 5%(0.05) 
t.test(Sales~Urban, data=df, alternative="two.sided", var.equal=TRUE)

# 유의 확률(0.759)가 유의수준보다 크기 때문에 귀무가설을 기각할 수 없다.
# 즉, 유의수준 0.05 하에서 Urban 변수에 따른 Sales의 차이는 없다.

```

```r
#------------------------------------------------------------------------------------------------
# Q2)Sales변수와 CompPrice, Income, Advertising, Population, Price, Age, Education 변수들 간에 
#    피어슨 상관계수를 이용한 상관관계 분석을 수행하고 이를 해석하시오.
#------------------------------------------------------------------------------------------------

str(df)

# method = One of "pearson" (default), "kendall", or "spearman"
# 피어슨 상관계수 확인
result <- cor(x = df[c("CompPrice", "Income", "Advertising", "Population", "Price", "Age", "Education")], y = df$Sales, method = "pearson", use = "pairwise.complete.obs")
print(result)

# CompPrice와 Education은 상관계수가 0에 가까워서 상관성을 거의 가지지 않음을 확인가능
# Price와 Sales는 약한 음의 상관과계가 있음을 확인 가능

# 만약 수치형 변수간의 모든 조합에 대한 상관도를 보려면 상관도 행렬을 아래처럼 수행
result <- cor(x = df[c("CompPrice", "Income", "Advertising", "Population", "Price", "Age", "Education")], method = "pearson", use = "pairwise.complete.obs")

pairs(result)
```

```r
#------------------------------------------------------------------------------------------------
# Q3) 종속변수를 Sales, 독립변수를 CompPrice, Income, Advertising, Population, Price, 
#     Age, Education으로 설정하고 후진제거법을 활용하여 회귀분석을 실시하고, 추정된 회귀식을 작성하시오. 
#------------------------------------------------------------------------------------------------

model <- lm(formula = Sales ~ CompPrice+Income+Advertising+Population+Price+Age+Education, data = df)

# 후진 제거법으로 변수 선택 및 회귀 분석 수행
model.pred <- step(model, direction = "backward")

# 선택된 변수
Sales ~ CompPrice + Income + Advertising + Price + Age

# 해당 변수로 회귀 분석 수행

model <- lm(formula = Sales ~ CompPrice + Income + Advertising + Price + Age, data = df)
summary(model)

# 아래 결과의 회귀 계수들이랑 변수들 합쳐서 추정 회귀식 적으면 됨
# lm(formula = Sales ~ CompPrice + Income + Advertising + Price + 
#     Age, data = df)

# Residuals:
#     Min      1Q  Median      3Q     Max 
# -4.9071 -1.3081 -0.1892  1.1495  4.6980 

# Coefficients:
#              Estimate Std. Error t value Pr(>|t|)    
# (Intercept)  7.109190   0.943940   7.531 3.46e-13 ***
# CompPrice    0.093904   0.007792  12.051  < 2e-16 ***
# Income       0.013092   0.003465   3.779 0.000182 ***
# Advertising  0.130611   0.014572   8.963  < 2e-16 ***
# Price       -0.092543   0.005044 -18.347  < 2e-16 ***
# Age         -0.044971   0.005994  -7.503 4.20e-13 ***
# ---
# Signif. codes:  0 '***' 0.001 '**' 0.01 '*' 0.05 '.' 0.1 ' ' 1

# Residual standard error: 1.927 on 394 degrees of freedom
# Multiple R-squared:  0.5403,	Adjusted R-squared:  0.5345 
# F-statistic: 92.62 on 5 and 394 DF,  p-value: < 2.2e-16

```
