```python
import numpy as np
import matplotlib.pyplot as plt

arr = np.arange(10)
print(arr)

# 단순히 직선 그래프 표시
# x축은 인덱스라고보면 되고, y축이 실제 numpy 데이터 값이라고 생각하면 됨
plt.plot(arr)

# 새로운 도화지를 꺼낸다는 의미로 figure 함수 사용
# fig만 생성하는걸로는 화면에 표시되는게 없고 반드시 서브 플롯 생성 필요
fig = plt.figure()
type(fig)

# 2X2의 서브플롯 중에서 첫번째(좌측상단) 그림을 선택한다는 것
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax4 = fig.add_subplot(2,2,4)

# 서브플롯 실컷 넣고나서 plot 호출하면, 가장 최근에 그려진 서브 플롯에 데이터 넣게됨
plt.plot([1,5,10,11])

# k--는 점선을 그리도록 하는 옵션
plt.plot(np.random.randn(50).cumsum(), 'k--')




```
