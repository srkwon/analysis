```r
library(dplyr)
library(caret)
library(rpart) 
library(e1071) 
library(class) 
library(nnet) 
library(randomForest) 


# 의사결정트리(결정트리, 결정나무) - rpart
# 서포트벡터머신 - e1071
# K-NN - class
# 인공신경망 - nnet
# 앙상블 - 랜덤포레스트 - randomForest

# 랜덤 코드 적용
set.seed(2021)

# 데이터 수집

data <- read.csv('./data/WA_Fn-UseC_-Telco-Customer-Churn.csv', stringsAsFactors = TRUE)

str(data)
summary(data)

# 데이터 전처리
# 8:2로 분할
#help(createDataPartition)
data.part <- createDataPartition(data$Churn, p = 0.8, )

data.train <- data[data.part$Resample1, ]
data.test <- data[-data.part$Resample1, ]

nrow(data.train); nrow(data.test)

# 모델 구축
# help(rpart)
model <- rpart(
        formula = Churn ~ .
        , data = data.train
)

# 모델 평가

model.pred <- predict(
    model
    , newdata = data.test
    , type = "class"
)

# str(model.pred)

confusionMatrix(
    data = model.pred # 예측값
    , reference = as.factor(data.test$Churn) # 실제값
)

# 
# help(write.csv)


write.csv(x = model.pred, "./data/20211128.csv", row.names = FALSE)
```

```r
library(dplyr)
library(caret)
library(rpart) 
library(e1071) 
library(class) 
library(nnet) 
library(randomForest) 

# 의사결정트리(결정트리, 결정나무) - rpart
# 서포트벡터머신 - e1071
# K-NN - class
# 인공신경망 - nnet
# 앙상블 - 랜덤포레스트 - randomForest

# 랜덤 코드 적용
set.seed(2021)

# 데이터 수집

data <- read.csv('./data/WA_Fn-UseC_-Telco-Customer-Churn.csv', stringsAsFactors = TRUE)

# 결측값은 다 없애고 이용
data <- data[complete.cases(data), ]

# str(data)
# summary(data)

# 데이터 전처리
# 8:2로 분할
#help(createDataPartition)
data.part <- createDataPartition(data$Churn, p = 0.8, )

data.train <- data[data.part$Resample1, ]
data.test <- data[-data.part$Resample1, ]

nrow(data.train); nrow(data.test)

# 의사결정트리

model.dt <- rpart(
        formula = Churn ~ .
        , data = data.train
)

model.dt.pred <- predict(
    model.dt
    , newdata = data.test
    , type = "class"
)

confusionMatrix(
    data = model.dt.pred # 예측값
    , reference = as.factor(data.test$Churn) # 실제값
)


# 서포트 벡터 머신
model.svm <- svm(
    formula = Churn ~ .
    , data = data.train
)

model.svm.pred <- predict(
    model.svm
    , newdata = data.test
    , type = "class"
)

confusionMatrix(
    data = model.svm.pred # 예측값
    , reference = as.factor(data.test$Churn) # 실제값
)

```

```r
library(dplyr)
library(caret)
library(rpart) 
library(e1071) 
library(class) 
library(nnet) 
library(randomForest) 

# 의사결정트리(결정트리, 결정나무) - rpart
# 서포트벡터머신 - e1071
# K-NN - class
# 인공신경망 - nnet
# 앙상블 - 랜덤포레스트 - randomForest

# 랜덤 코드 적용
set.seed(2021)

# 데이터 수집

data <- read.csv('./data/WA_Fn-UseC_-Telco-Customer-Churn.csv', stringsAsFactors = FALSE)

# 결측값은 다 없애고 이용
data <- data[complete.cases(data), ]

# str(data)
# summary(data)

# 데이터 전처리
# 8:2로 분할
#help(createDataPartition)
data.part <- createDataPartition(data$Churn, p = 0.8, )

data.train <- data[data.part$Resample1, ]
data.test <- data[-data.part$Resample1, ]

nrow(data.train); nrow(data.test)

# KNN
# help(knn)

# knn.model <- knn(
#     train = data.train
#     , cl = data.train$Churn
#     , test = data.test
#     , k = 1
# )

########################################################################################################
# 랜덤포레스트
########################################################################################################

# str(data)

model.rf <- randomForest(
    formula = as.factor(Churn) ~ .
    , data = data.train
)

print(model.rf)

# 변수 중요도 확인 가능, 큰 값일수록 중요한 변수
importance(model.rf)

model.rf.pred <- predict(
    model.rf
    , newdata <- data.test
)

confusionMatrix(
    data = model.rf.pred # 예측값
    , reference = as.factor(data.test$Churn) # 실제값
)

########################################################################################################
# ANN
########################################################################################################

# help(nnet)
model.ann <- nnet(
    formula = as.factor(Churn) ~ gender+Partner+MonthlyCharges
    , data = data.train
    , size = 2 # 은닉층 개수
    , maxit = 200 # 반복 횟수
)

model.ann.pred <- predict(
    model.ann
    , newdata <- data.test
)

confusionMatrix(
    data = as.factor(ifelse(model.ann.pred > 0.5, 'Yes', 'No')) # 예측값
    , reference = as.factor(data.test$Churn) # 실제값
)
```
