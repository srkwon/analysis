```r
library(dplyr)
data(BostonHousing)

data <- BostonHousing

str(data)
summary(data)

# crim의 상위 10번째 값으로 상위 10개값 변환
crim.10th <- data %>%
    arrange(desc(crim)) %>%
    head(10)

data$crim <- ifelse(data$crim >= crim.10th$crim[10], crim.10th$crim[10], data$crim)

# age 80 이상인 값에 대해 crim 평균계산

crim.avg = data %>%
    filter(age >= 80) %>%
    summarise(crim.avg = mean(crim))

print(crim.avg)   

```

```r
library(dplyr)

data <- read.csv('./data/insurance.csv')

str(data)
summary(data)

charges.mean <- mean(data$charges, na.rm = TRUE)
charges.sd <- sd(data$charges, na.rm = TRUE)

charges.sum <- data %>% 
    filter((charges >= charges.mean + 1.5 * charges.sd) | (charges <= charges.mean - 1.5 * charges.sd)) %>%
    summarise(sum(charges))

print(charges.sum)

```

```r
library(dplyr)
library(caret)
library(rpart) # dt
library(e1071) # svm
library(ipred) # bagging
library(xgboost) # boosting

set.seed(20211204)

data <- iris
str(data)
summary(data)

# 데이터 전처리
part <- createDataPartition(
    data$Species
    , p = 0.7
)

data.train <- data[part$Resample1, ]
data.test <- data[-part$Resample1, ]

nrow(data.train) ; nrow(data.test)


# 모형 구축
## rpart
model.rpart <- rpart(
    formula = Species ~ .
    , data = data.train
)

# ?predict

model.rpart.pred <- predict(
    model.rpart
    , newdata = data.test
    , type="class" # class or response
)

# 0.9556
confusionMatrix(
    data = model.rpart.pred
    , reference = data.test$Species
)

## svm
model.svm <- svm(
    formula = Species ~ .
    , data = data.train
)

model.svm.pred <- predict(
    model.svm
    , newdata = data.test
    , type = "class"
)

# 0.9778
confusionMatrix(
    data = model.svm.pred
    , reference = data.test$Species
)


model.bagging <- bagging(
    formula = Species ~ .
    , data = data.train
)

model.bagging.pred <- predict(
    model.bagging
    , newdata = data.test
    , type = "class"
)

confusionMatrix(
    data = model.bagging.pred
    , reference = data.test$Species
)




# SVM 선택 및 파일 생성
write.table(model.svm.pred, col.names = c("TEST") ,file = './data/20211202.csv', row.names = FALSE)

help(write.csv)
```

```r
library(class)
library(dplyr)
library(caret)

data <- iris
str(data)
summary(data)

set.seed(2021)

part <- createDataPartition(
    data$Species
    , p = 0.7
)

data.train <- data[part$Resample1, ]
data.test <- data[-part$Resample1, ]

nrow(data.train);nrow(data.test)

#knn

# ?knn

# knn은 중요한 것이, 데이터셋에 숫자가 아닌 값이 들어가면 안됨, 명심할 것
model.knn.k1 <- knn(
    train = data.train[, -5]
    , test = data.test[, -5]
    , cl = data.train$Species
    , k = 1
)

confusionMatrix(
    data = model.knn.k1
    , reference = data.test$Species
)

model.knn.k2 <- knn(
    train = data.train[, -5]
    , test = data.test[, -5]
    , cl = data.train$Species
    , k = 2
)

confusionMatrix(
    data = model.knn.k2
    , reference = data.test$Species
)

model.knn.k4 <- knn(
    train = data.train[, -5]
    , test = data.test[, -5]
    , cl = data.train$Species
    , k = 4
)

confusionMatrix(
    data = model.knn.k4
    , reference = data.test$Species
)

model.knn.k8 <- knn(
    train = data.train[, -5]
    , test = data.test[, -5]
    , cl = data.train$Species
    , k = 8
)

confusionMatrix(
    data = model.knn.k8
    , reference = data.test$Species
)

# k = 4로 선정
```

```r
library(dplyr)
library(caret)
library(e1071)

data <- read.csv('./data/Loan payments data.csv', stringsAsFactors = TRUE)

str(data)
summary(data)

head(data)

table(data$loan_status)

# 데이터 전처리

data$past_due_days = ifelse(is.na(data$past_due_days), median(data$past_due_days, na.rm = TRUE), data$past_due_days)

summary(data)

part <- createDataPartition(data$loan_status, p = 0.7)

data.train <- data[part$Resample1, ]
data.test <- data[-part$Resample1, ]

nrow(data.train);nrow(data.test)

# svm

# help(svm)
model.svm <- svm(
    formula = loan_status ~ .
    , data = data.train
    , type = "C-classification"
)

model.svm.pred <- predict(
    model.svm
    , newdata = data.test
)

length(model.svm.pred)
length(data.test$loan_status)

confusionMatrix(
    data = model.svm.pred
    , reference = factor(data.test$loan_status)
)

# 정확도 0.6

nrow(data.test)
mode(model.svm.pred)

result <- cbind(data.test$Loan_ID, as.character(model.svm.pred))
colnames(result) <- c("Loan_ID", "loan_status")

write.csv(result, './data/20211203-2.csv', row.names = FALSE)

?factor
```
