```r
library(dplyr)

min.max.scale <- function(x) {
    return (x - min(x)) / (max(x) - min(x))
}

data <- mtcars

str(data)

data.filtered <- data %>%
    select(wt) %>%
    mutate(wt.scaled = min.max.scale(wt)) %>%
    filter(wt.scaled > 0.5)

print(count(data.filtered))
```

```r
library(dplyr)
library(caret)

library(rpart)
library(e1071)

data(iris)

# 데이터 수집
data <- iris

str(data)

# 데이터 전처리

set.seed(20211129)
part <- createDataPartition(
    data$Species
    , p = 0.7
)

data.train <- data[part$Resample1,]
data.test <- data[-part$Resample1,]

nrow(data.train);nrow(data.test)

## rpart

# ?rpart

model.rpart <- rpart(
    formula = as.factor(Species) ~ .
    , data = data.train
)

# type = class => 범주
# type = respnse => 수치
model.rpart.pred <- predict(
    model.rpart
    , newdata = data.test
    , type = "class"
)

model.rpart.pred

confusionMatrix(
    data = model.rpart.pred
    , reference = data.test$Species
)

## SVM 체크

model.svm <- svm(
    formula = as.factor(Species) ~ .
    , data = data.train
)

# type = class => 범주
# type = respnse => 수치
model.svm.pred <- predict(
    model.svm
    , newdata = data.test
    , type = "class"
)

confusionMatrix(
    data = model.svm.pred
    , reference = data.test$Species
)

# SVM의 accuracy가 더 높으므로, SVM 모델로 예측한 결과 값을 파일로 작성
write.csv(model.svm.pred, "./data/20211129.csv")
```
